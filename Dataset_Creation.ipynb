```python
links = []
for i in range(1,151):
    url = f"https://www.onlinekhabar.com/content/news/page/{i}"
    
    homepage = requests.get(url)
    contents = BeautifulSoup(homepage.content, 'html.parser')
    for news in contents.findAll('div', class_='ok-news-post'):
        links.append(news.a['href'])
links[:5]
```

```python
from tqdm import tqdm
```

```python
news = []


for link in tqdm(links):
    page = requests.get(link)
    contents = BeautifulSoup(page.content, 'html.parser')
    for titles in contents.findAll('h1'):
        title = titles.text
        title = title.replace(u'\xa0', u' ')
        title = title.replace(u'\n', u' ')

    for articles in contents.findAll('div', class_='ok18-single-post-content-wrap'):
        article = articles.text
        article = article.replace(u'\xa0', u' ')
        article = article.replace(u'\n', u' ')

    news.append([link, title, article])
```

```python
news = pd.DataFrame(news, columns = ['link', 'title', 'article'])
news
```

```python
news.drop_duplicates(subset=['title'], inplace=True)
```

```python
news = news[['title', 'article','link']]
news.insert(0, 'id', range(0, news.shape[0]))
```
