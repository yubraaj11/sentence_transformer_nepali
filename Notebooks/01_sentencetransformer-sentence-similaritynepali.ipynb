```python
def translate_text(text, target_lang):
    translator = GoogleTranslator(source="auto", target=target_lang)
    translation = translator.translate(text)

    return translation


target_lang = "ne"
columns_to_translate = ["sentence1", "sentence2"]

for column in columns_to_translate:
    tqdm.pandas(desc=f"Translating {column}")
    df[column] = df[column].progress_apply(lambda x: translate_text(x, target_lang))
```

```python
import re

def remove_alphabets(text):
    if isinstance(text, str):
        pattern = r'[a-zA-Z]'
        return re.sub(pattern, '', text)
    else:
        return text

df = df.applymap(remove_alphabets)
```

# 2. Training Sentence Transformer

```python
!pip install sentence_transformers
```

```python
import torch
from torch.utils.data import DataLoader
import math
from sentence_transformers import SentenceTransformer,  LoggingHandler, losses, models, util
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator
from sentence_transformers.readers import InputExample
import logging
from datetime import datetime
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model_name = 'Rajan/NepaliBERT'

train_batch_size = 16
num_epochs = 8
model_save_path = 'output/sentence_transformer_nepali'
```

```python
word_embedding_model = models.Transformer(model_name)
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension(), 
                               pooling_mode_mean_tokens=True,
                               pooling_mode_cls_token=False,
                               pooling_mode_max_tokens=False,
                               )
```

```python
model = SentenceTransformer(modules=[word_embedding_model, pooling_model])
model.to(device)
```

```python
input_example_samples = []

for index, row in df.iterrows():
    score = float(row['score']) / 5.0 # Normalize score between 0 to 1
    inp_example = InputExample(texts=[row['sentence1'], row['sentence2']], label=score)

    input_example_samples.append(inp_example)
```

```python
len(input_example_samples)
```

```python
import random

random.shuffle(input_example_samples)

train_ratio = 0.8
test_ratio = 0.1
dev_ratio = 0.1

num_examples = len(input_example_samples)
num_train = int(num_examples * train_ratio)
num_dev = int(num_examples * dev_ratio)
num_test = int(num_examples * test_ratio)


train_samples = input_example_samples[:num_train]
dev_samples = input_example_samples[num_train:num_train + num_dev]
test_samples = input_example_samples[num_train + num_dev:]


```

```python
print("Train samples:", len(train_samples))
print("Dev samples:", len(dev_samples))
print("Test samples:", len(test_samples))
```

```python
train_dataloader = DataLoader(train_samples, shuffle=True, batch_size = train_batch_size)
train_loss = losses.CosineSimilarityLoss(model=model)
```

```python
evaluator = EmbeddingSimilarityEvaluator.from_input_examples(dev_samples, name='stsb-dev-nepali')
```

```python
warmup_steps = math.ceil(len(train_dataloader) * num_epochs * 0.1) # 10% of train data for warm-up
```

```python
model.fit(train_objectives=[(train_dataloader, train_loss)],
          evaluator = evaluator,
          epochs = num_epochs,
          evaluation_steps = 1000,
          warmup_steps = warmup_steps,
          output_path = model_save_path
)
```
